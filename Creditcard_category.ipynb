# count of fraudulent transactions from dataset

```python
print(df['is_fraud'].value_counts())
```

# filter  data to count number of fraud cases in each category

```python

fraud_cases =df[df['is_fraud']==1]
category_cases = fraud_cases.groupby('category')['is_fraud'].count().sort_values(ascending=False)
print(category_cases)
```

# plot a chart showing the displaying highest to lowest occuring fraud in each category. 

```python
cmap = plt.get_cmap('viridis')
colors = [cmap(i/len(category_cases)) for i in range(len(category_cases))]
category_cases.plot(kind='bar', color=colors)




plt.title('category with their  number of fraudulent transactions')
plt.xlabel('category')
plt.ylabel('Number of fraud')
plt.show()
```
